```python
import pandas
```

```python
names = ['11', '21', '22', '23', '31', '42', '44', '48', '51', '52', '53', '54', '55', '56', '61', '62', '71', '72', '81', '92']
```

```python
def get_correlation(df1, df2):
    common_cols = [col for col in df1.columns if col in df2.columns and col not in ['metrics']]
    s1 = df1[common_cols].iloc[0].astype(float)
    s2 = df2[common_cols].iloc[0].astype(float)
    return s1.corr(s2)
```

```python
d1_us_corel = 0
count = 0
for name in names:
    d1_data = pandas.read_csv(f"../data.buggy/paper/d2/data/distance/{name}.csv")
    us_data = pandas.read_csv(f"/home/amiri/Downloads/distance/{name}.csv")
    d1_data = d1_data[d1_data['metrics'] == 'mean']
    us_data = us_data[us_data['metrics'] == 'mean']
    count += 1
    correlation = get_correlation(d1_data, us_data)
    d1_us_corel += correlation
    print(f"Corel for sector {name}: {correlation}")
print(f"Average correlation: {d1_us_corel / count}")
```
